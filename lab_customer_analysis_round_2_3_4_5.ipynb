```python
print('Maximum value',File2['Total Claim Amount'].max())
print('Maximum value',File2['Total Claim Amount'].min())
print('Mean value',File2['Total Claim Amount'].mean())
```

```python
def tca_bins(x):
    if x <=450:
        return 1
    elif x<=900:
        return 2
    elif x<=1350:
        return 3
    elif x<=1800:
        return 4
    else:
        return 5
```

```python
File2['Total Claim Amount bins']=list(map(tca_bins, File2['Total Claim Amount']))
File2
```

```python
File2.groupby('Total Claim Amount bins')['Response_rate'].mean().plot.bar()
plt.show()
```

```python
sns.barplot(x='Total Claim Amount bins', y='Response_rate', data=File2)
```

```python
fig, ax=plt.subplots()
sns.barplot(x='Total Claim Amount bins', y='Response_rate', data=File2)
labels=['Under 450', '450-900','900-1350','1350-1800','Over 1800']
ax.set_xticklabels(x='Total Claim Amount bins', labels=labels,rotation='vertical')
plt.show()
```
